### ANAC Aeródromos Cadastrados <a href="http://www.anac.gov.br/assuntos/setor-regulado/aerodromos/cadastro-de-aerodromos/aerodromos-cadastrados">fonte<a/>

```python
# helper function
from os import listdir
def dir_files(name: str, folder: str):
    files = [folder + e for e in listdir(folder) if (e[-4:] == ".xls") or (e[-4:] == ".csv")]
    print(name+": "+ ", ".join(files))
    return files
```

```python
only_civil = dir_files('Aeródromos civis','only_civil/')
```

```python
public_private = dir_files('Aeródromos públicos e privados', 'public_private/')
```

### Aeródromos civis

```python
from pandas import read_csv
```

```python
df1 =  read_csv(only_civil[0], sep=';', header=[0,1], encoding='latin-1')
df1.head()
```

```python
df2 =  read_csv(only_civil[1], sep=';', header=[0,1], encoding='latin-1')
df2.head()
```

### Aeródromos Privados

```python
from pandas import read_excel
```

```python
df3 =  read_excel(public_private[0], sheet_name=0, header=[0,1])
df3.head()
```

### Aeródromos Publicos

```python
df4 =  read_excel(public_private[1], sheet_name=0, header=[1,2])
df4.head()
```

### Gerando lista de ICAOs únicos para arquivo *.TXT

```python
from pandas import concat
icao_list = concat([df1.iloc[:,0], df2.iloc[:,0], df3.iloc[:,0], df4.iloc[:,0]])
```

```python
icao_list.drop_duplicates(inplace=True)
```

```python
icao_list.head()
```

```python
icao_list.to_csv('brazilian_aerodromes.icao', header=False, index=None, sep=' ', mode='a')
```
